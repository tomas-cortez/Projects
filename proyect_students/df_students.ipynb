### Importamos el archivo 'Alu_Carreras.csv' y realizamos sus respectivas transformaciones

```python
alumnos=pd.read_csv('Alu_Carreras.csv',sep=";",encoding="latin",engine="python")
alumnos.head()
```

```python
alumnos.dtypes
```

```python
alumnos["Fecha_Nac"]=pd.to_datetime(alumnos["Fecha_Nac"])
alumnos.dtypes
alumnos['periodo_ingreso'] = "Año "+alumnos['Año_Ingreso'].astype(str) + " / " + alumnos['Semestre_Ingreso'].astype(str)+"° Semestre"
alumnos
```

```python
alumnos["Fecha_Nac"]=pd.to_datetime(alumnos["Fecha_Nac"])
alumnos['mes'] = pd.DatetimeIndex(alumnos['Fecha_Nac']).month
alumnos['year'] = pd.DatetimeIndex(alumnos['Fecha_Nac']).year
alumnos['Edad']=alumnos['Año_Ingreso']- alumnos['year']
alumnos['Edad']=alumnos['Edad'].astype(float)
alumnos
```

### Generamos las columnas nesesarias para su posterior analisis 

```python
alumnos.loc[(alumnos['Semestre_Ingreso'] ==1) & (alumnos['mes']>=3.0), 'Edad'] =alumnos['Edad']-1  
alumnos.loc[(alumnos['Semestre_Ingreso'] ==2) & (alumnos['mes']>=8.0), 'Edad'] =alumnos['Edad']-1  
alumnos=alumnos[(alumnos.Año_Ingreso >=2000)]
alumnos=alumnos[(alumnos.Edad>=17)]
alumnos
```

### Importamos el archivo 'Alu_Materias.csv' y realizamos sus respectivas transformaciones

```python
materias=pd.read_csv('Alu_Materias.csv',sep=";",encoding="latin",engine="python")
materias.loc[materias.Nota1== 11 , 'Nota1'] = -1
materias.loc[materias.Nota2== 11 , 'Nota2'] = -1
materias.loc[materias.Nota3== 11 , 'Nota3'] = -1
materias.loc[materias.Recup== 11 , 'Recup'] = -1
materias['Periodo_ingreso'] = "Año "+materias['Año'].astype(str) + " / " + materias['Semestre'].astype(str)+"° Semestre"

#materias.loc[materias.Nota1.isnull(),'Nota1'] = -1
#materias.loc[materias.Nota2.isnull(),'Nota2'] = -1
#materias.loc[materias.Nota3.isnull(),'Nota3'] = -1
#materias.loc[materias.Recup.isnull() & ((materias['Nota1']<5) | (materias['Nota2']<5)),'Recup'] = -1
#
materias.head()
```

### Generamos una columna con informacion sobre la 'Condicion' del alumno con año mayor a 2005.
### Esta puede contener los siguientes valores
- Promocionado
- Regular
- Desertor
- Libre
- Sin_Datos        


```python
matcopy = materias[(materias["Año"]>2005)] 
```

```python
conditionlist2 = [
(matcopy['Nota1'] >= 7.0) & (matcopy['Nota2'] >= 7.0) & (matcopy['Nota3'] >= 7.0) ,
((matcopy['Nota1'] >=4) & (matcopy['Nota2']>=4) & (matcopy['Nota3']>=4)) | (matcopy['Recup']>=4) ,
( ( ( (matcopy['Nota1']<=4) | (matcopy['Nota1']==-1) ) | ( (matcopy['Nota2']<=4) | (matcopy['Nota2']==-1) ) | ( (matcopy['Nota3']<=4) | (matcopy['Nota3']==-1) ) )  &  (matcopy['Recup'] == -1)),
(matcopy['Nota3']==-1),
(matcopy["Nota1"].isnull()) & (matcopy["Nota2"].isnull()) & (matcopy["Nota3"].isnull()) & (matcopy["Recup"].isnull()) 
]
choicelist2 = ['Promocionado', 'Regular', 'Desertor','Desertor',"Sin_Datos"]
matcopy['Condicion'] = np.select(conditionlist2, choicelist2, default="Libre")
```

```python
can=matcopy.Condicion.value_counts()
can
```

### Generamos una columna con informacion sobre la 'Condicion' del alumno con año menor a 2005.
### Esta puede contener los siguientes valores
- Promocionado
- Regular
- Desertor
- Libre
- Sin_Datos        

```python
matcopy2 = materias[(materias["Año"]<=2005)] 
```

```python
conditionlist3 = [
(matcopy2['Nota1'] >= 7.0) & (matcopy2['Nota2'] >= 7.0) ,
( ( (matcopy2['Nota1'] >=4) & (matcopy2['Nota2']>=4) ) | (matcopy2['Recup']>=4)) ,
((((matcopy2['Nota1']<4) | (matcopy2['Nota1']==-1)) & ((matcopy2['Nota2']<4) | (matcopy2['Nota2']==-1))) &  (matcopy2['Recup'] == -1)),
(matcopy2["Nota1"].isnull()) & (matcopy2["Nota2"].isnull()) & (matcopy2["Recup"].isnull()) 
]
choicelist3 = ['Promocionado', 'Regular', 'Desertor',"Sin_Datos"]
matcopy2['Condicion'] = np.select(conditionlist3, choicelist3, default="Libre")
```

```python
can1=matcopy2.Condicion.value_counts()
can1
```

# Ejercicio 2 SVM

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('data/winequality-white.csv', sep=';')
df.head()
```

Pronosticar quality, problema de clasificación. 

```python
from sklearn.model_selection import train_test_split

X=df.drop(axis=1,columns='quality')
y=df['quality']

#Train / Test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=123)

# Validation
Xval_train,Xval_test,yval_train,yval_test=train_test_split(X_train,y_train,test_size=0.20,random_state=123)
```

```python
from sklearn.svm import SVC
```

## SVM Lineal

```python
for c in [0.01,0.1,1,10,100]    
    
    svm=SVC(C=c, kernel='linear')
    svm.fit(Xval_train,yval_train)
    acc=svm.score(Xval_test,yval_test)

    print("Kernel:linear")
    print("C: ",c)
    print("Accuracy: ",acc)
```

## SVM polinómico grado 2

```python
for c in [10,100,1000, 10000]:    
    
    svm=SVC(C=c, kernel='poly',degree=2)
    svm.fit(Xval_train,yval_train)
    acc=svm.score(Xval_test,yval_test)
    
    print("--------------------------------")
    print("Kernel:poly, grado 2")
    print("C: ",c)
    print("Accuracy: ",acc)
```

## Modelo Seleccionado

## Evaluación del modelo  elegido

En train / test

```python
svm_eval=SVC(C=1, kernel='linear',probability=True)
svm_eval.fit(X_train,y_train)
acc=svm_eval.score(X_test,y_test)

print("Kernel:linear")
print("C: ",1)
print("Accuracy: ",acc)
```

```python
# Si queremos ver las probabilidades para cada clase:
np.set_printoptions(threshold=np.inf)
np.round(svm_eval.predict_proba(X_train),3)
```
